## 数据处理&清洗

```python
import pandas as pd
import numpy as np

train = pd.read_csv('/tmp/happiness/happiness_train_complete.csv',encoding='ISO-8859-1')
test = pd.read_csv('/tmp/happiness/happiness_test_complete.csv',encoding='ISO-8859-1')
```

```python
train.isna().sum()
```

```python
#查看label分布
y_train_=train["happiness"]
y_train_.value_counts()
```

```python
#index改为 level_0 从零开始
train.reset_index(inplace=True,drop = True)
```

```python
train.head()
```

```python
# happiness 为 -8 纯属是瞎填的 改为说不上幸福
y_train_=y_train_.map(lambda x:3 if x==-8 else x)
y_train_.value_counts()
```

```python
#将test和train联合起来处理删除部分不必要或不好的数据
#https://zhuanlan.zhihu.com/p/132593960
#忽略掉两者都没有的部分
data_union = pd.concat([train,test],axis=0,ignore_index=True)
```

```python
data_union.shape
```

```python
#edu_status
data_union = data_union.drop(['edu_other'],axis = 1)
data_union = data_union.drop(['join_party'],axis = 1)
data_union = data_union.drop(['survey_time'],axis = 1)
data_union = data_union.drop(['invest_other'],axis = 1)
data_union = data_union.drop(['property_other'],axis = 1)

#像有些 比如 s_edu 就比较让人头疼 怎么说呢 直接归为其他可以吗 我觉得是可以的 这个很难以判断
#再仔细观察 发现1718可能是没有配偶的情况
```
